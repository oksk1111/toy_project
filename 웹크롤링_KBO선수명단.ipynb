```
import requests
from bs4 import BeautifulSoup
```

```
source = requests.get("https://www.koreabaseball.com/Player/RegisterAll.aspx").text
soup = BeautifulSoup(source, "html.parser")

# 팀명
hotKeys = soup.select("th.fir")

index = 0
for key in hotKeys:
  print(key.text)
```

```
# 선수명
hotKeys = soup.select("table[class='tData tDays'] tbody li")

for key in hotKeys:
  print(key.text.split('(')[0].rstrip())
```
